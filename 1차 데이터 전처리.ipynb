# 데이터 불러오기 & 합치기

```python
import pandas as pd
import numpy as np

# Raw Data 불러오기, encoding = 'cp949' 컬럼명을 한글로 불러오기, low_memory=False 데이터 통째로 불러오기
df_2018 = pd.read_csv("2018.csv", encoding='cp949', low_memory=False)
df_2019 = pd.read_csv("2019.csv", encoding='cp949', low_memory=False)
df_2020 = pd.read_csv("2020.csv", encoding='cp949', low_memory=False)
df_2021 = pd.read_csv("2021.csv", encoding='cp949', low_memory=False)
df_2022 = pd.read_csv("2022.csv", encoding='cp949', low_memory=False)
df_2023 = pd.read_csv("2023.csv", encoding='cp949', low_memory=False)
df_2024 = pd.read_csv("2024.csv", encoding='cp949', low_memory=False)

# 컬럼 확인
pd.set_option('display.max_columns', None) # 잘리는 컬럼 없이 다 출력
display(df_2018.head(5))
display(df_2019.head(5))
display(df_2020.head(5))
display(df_2021.head(5))
display(df_2022.head(5))
display(df_2023.head(5))
display(df_2024.head(5))
```

```python
# 7개년 데이터 합치기
df_2018_2024 = pd.concat([df_2018, df_2019, df_2020, df_2021, df_2022, df_2023, df_2024], axis=0, ignore_index=True)

# concat 테이블 데이터 확인
display(df_2018_2024)

# concat 테이블 정보 확인
df_2018_2024.info()
```

# 프로젝트 목적에 맞는 컬럼 선정

```python
# 필요 없는 컬럼 제거
df_2018_2024 = df_2018_2024.drop(['접수연도', '자치구코드', '법정동코드', '지번구분', '지번구분명', '토지면적(㎡)', '권리구분', '신고구분', '신고한 개업공인중개사 시군구명'], axis=1)
df_2018_2024.info()
```

# 데이터 필터링

```python
# 아파트로 필터링
df_2018_2024 = df_2018_2024[df_2018_2024['건물용도'] == '아파트'].copy()
df_2018_2024.info()
```

```python
# 계약 연도 2017년부터 필터링 (가계약 고려)
df_2018_2024 = df_2018_2024[df_2018_2024['계약일'] >= 20170101].copy()
df_2018_2024.info()
```

# 결측치 처리

```python
# 본번이 null이 아닌 애들로 필터링
df_2018_2024 = df_2018_2024[df_2018_2024['본번'].notna()].copy()
df_2018_2024.info()
```

```python
# 건축년도 null 필터링
df_2018_2024 = df_2018_2024[df_2018_2024['건축년도'].notna()].copy()
df_2018_2024.info()
```

```python
# 자치구명 null 필터링
df_2018_2024 = df_2018_2024[df_2018_2024['자치구명'].notna()].copy()
df_2018_2024.info()
```

# 이상치 처리

```python
# 건축년도 0 필터링
df_2018_2024 = df_2018_2024[df_2018_2024['건축년도'] != 0.0].copy()
df_2018_2024.info()
```

```python
# 본번 int화
df_2018_2024['본번'] = df_2018_2024['본번'].astype(int)

# 본번이 0이 아닌 애들로 필터링
df_2018_2024 = df_2018_2024[df_2018_2024['본번'] != 0].copy()
df_2018_2024.info()
```

```python
# 건물면적 0 필터링
df_2018_2024 = df_2018_2024[df_2018_2024['건물면적(㎡)'] != 0.0].copy()
df_2018_2024.info()
```

```python
# 물건금액 0 필터링
df_2018_2024 = df_2018_2024[df_2018_2024['물건금액(만원)'] != 0].copy()
df_2018_2024.info()
```

```python
# 취소일이 null 필터링
df_2018_2024 = df_2018_2024[df_2018_2024['취소일'].isnull()].copy()
df_2018_2024.info()
```

```python
# 아파트 층수가 1층 이상(지상)
df_2018_2024 = df_2018_2024[df_2018_2024['층'] > 0].copy()
df_2018_2024.info()
```

```python
total_len = len(df_2018_2024)
dup_len = df_2018_2024.duplicated().sum()
print(f"전체 데이터: {total_len}건")
print(f"중복 데이터: {dup_len}건 ({dup_len/total_len*100:.2f}%)")
```

# 파생변수 생성

```python
# 계약일 -> 계약 연도 컬럼 추가(int 타입)
df_2018_2024['계약 연도'] = df_2018_2024['계약일'] // 10000

# 계약일 컬럼 제거
df_2018_2024 = df_2018_2024.drop('계약일', axis=1)

df_2018_2024.info()
```

```python
# 평수 컬럼 추가(정수까지 반올림해서 int화)
df_2018_2024['평수(평)'] = round(df_2018_2024['건물면적(㎡)'] / 3.3)
df_2018_2024['평수(평)'] = df_2018_2024['평수(평)'].astype(int)
# df_2018_2024.info()

# 평당가 계산(정수까지 반올림해서 int화)
df_2018_2024['평당가(만원)'] = round(df_2018_2024['물건금액(만원)']/ df_2018_2024['평수(평)'])
df_2018_2024['평당가(만원)'] = df_2018_2024['평당가(만원)'].astype(int)
df_2018_2024.info()
```

```python
# # 시명 : 서울특별시 추가
df_2018_2024['시명'] = "서울특별시"

# 부번 정수화
df_2018_2024['부번'] = df_2018_2024['부번'].astype(int)

# # 부번 0 -> Null 처리
df_2018_2024['부번'] = df_2018_2024['부번'].replace(0, None)

df_2018_2024.info()
```

```python
# 주소 컬럼 생성
main_num = df_2018_2024['본번'].fillna(0).astype(float).astype(int).astype(str)
sub_num = df_2018_2024['부번'].fillna(0).astype(float).astype(int).astype(str)

hyphen_str = np.where((df_2018_2024['부번'].isna()) | (df_2018_2024['부번'] == 0), "", "-" + sub_num)

df_2018_2024['주소'] = (df_2018_2024['시명'].fillna('') + ' ' + df_2018_2024['자치구명'].fillna('') + ' ' + df_2018_2024['법정동명'].fillna('') + ' ' + main_num + hyphen_str)
df_2018_2024.info()
```

```python
# 층, 건축년도 데이터 정수화
df_2018_2024['층'] = df_2018_2024['층'].astype(int)
df_2018_2024['건축년도'] = df_2018_2024['건축년도'].astype(int)
df_2018_2024.info()
```

```python
# 최종 정렬
df_2018_2024 = df_2018_2024[['계약 연도', '시명', '자치구명', '법정동명', '본번', '부번', '주소', '건물명', '물건금액(만원)', '건물면적(㎡)', '평당가(만원)', '평수(평)', '층', '취소일', '건축년도', '건물용도']]

# 인덱스 번호 정렬
df_2018_2024 = df_2018_2024.reset_index(drop=True)

# 1차 전처리 결과 csv로 내보내기
df_2018_2024.to_csv('1차 데이터 전처리.csv', encoding='cp949', index=False)
```
